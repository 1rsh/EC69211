#### Writing Image

```python
from utils import convolve, show_modified

kernel1 = [[0, -1, 0], 
          [-1, 5, -1], 
          [0, -1, 0]]

kernel2 = [[1, 0, -1], 
          [1, 0, -1], 
          [1, 0, -1]]
```

```python
bmp = BMP()
bmp.read("cameraman.bmp")
img_to_save1 = convolve(bmp.img, kernel2)
print(f"Shape of img array: ({len(img_to_save1)}, {len(img_to_save1[0])}) \nValue at (100, 100): {img_to_save1[100][100]}")
show_modified(bmp.img, img_to_save1)
```

```python
bmp = BMP()
bmp.read("corn.bmp")

orig_img = [[None] * bmp.shape[1] for _ in range(bmp.shape[0])]

for i in range(bmp.shape[0]):
    for j in range(bmp.shape[1]):
        orig_img[i][j] = [bmp.raw_img[i][j]] * 3
        
img_to_save2 = convolve(orig_img, kernel1)
color_map = bmp.color_map
print(f"Shape of img array: ({len(img_to_save2)}, {len(img_to_save2[0])}) \nValue at (100, 100): {img_to_save2[100][100]}")
show_modified(orig_img, img_to_save2)
```

```python
bmp = BMP()
bmp.read("pepper.bmp")
img_to_save3 = convolve(bmp.img, kernel2)
print(f"Shape of img array: ({len(img_to_save3)}, {len(img_to_save3[0])}) \nValue at (100, 100): {img_to_save3[100][100]}")
show_modified(bmp.img, img_to_save3)
```

```python
savepath = 'save_cameraman.bmp'
bmp = BMP()
bmp.write(savepath, img_to_save1)
```

```python
savepath = 'save_corn.bmp'
bmp = BMP()

img_to_save2_grey = [[None] * len(img_to_save2[0]) for _ in range(len(img_to_save2))]

for i in range(len(img_to_save2)):
    for j in range(len(img_to_save2[0])):
        img_to_save2_grey[i][j] = img_to_save2[i][j][0]

totally_random = [(random.randint(0, 255), random.randint(0, 255), random.randint(0, 255)) for _ in color_map]
random_gradient = [tuple(int(totally_random[0][j] + i/255 * (totally_random[-1][j] - totally_random[0][j])) for j in range(3)) for i in range(256)]

bmp.write(savepath, img_to_save2_grey, color_map = random_gradient)
```

```python
savepath = 'save_pepper.bmp'
bmp = BMP()
bmp.write(savepath, img_to_save3)
```
